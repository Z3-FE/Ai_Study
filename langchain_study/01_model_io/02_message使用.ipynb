# 3.模型调用的方法
ChatPromptTemplate 提示词模版，
📜 Runnable 协议详解
Runnable 协议的主要目标是 简化和标准化 LangChain 中各个组件（如 LLM、提示模板、解析器、工具等）的调用和组合方式，使得构建复杂的 AI 链条（Chains）和表达式语言（LCEL）变得非常容易。

1. 核心概念和作用
标准接口： 它为所有 LangChain 组件（称为 Runnable 对象）提供了一致的输入和输出接口。无论是调用一个简单的 LLM 模型，还是一个复杂的 Agent，调用的方法都是一样的。

链式组合 (Chaining)： 它是 LangChain 表达式语言 (LCEL) 的基础。通过使用 | 运算符，您可以像管道一样将不同的 Runnable 组件连接起来，构建数据流。

例如：prompt | llm | parser 就是一个典型的 Runnable 链。

统一调用方法： 所有实现 Runnable 协议的对象都必须提供以下核心方法：

invoke()：以同步方式（Sync）调用组件，并等待结果返回。

ainvoke()：以异步方式（Async）调用组件。

batch()：同时处理多个输入（批量调用）。

abatch()：异步批量调用。

stream()：流式返回 响应块（例如，实时显示 LLM 生成的文本）。

astream()：异步流式返回。

```python
import asyncio
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
# (假设您的 LLM 定义已就位)

# 定义 Prompt Template  提示词模版
prompt = ChatPromptTemplate.from_template("请用一句话专业地解释什么是 {topic}。")

# 组合 Chain
# Prompt 接收字典 -> LLM 接收格式化消息 -> Parser 提取纯文本
chain = prompt | llm | StrOutputParser() # <--- 定义完整的 Chain

print('chain:',chain)
# --- 核心调用模式示例 ---

## 模式 1: 同步调用 (invoke)
async def example_invoke():
    """
    同步调用：最简单、最常用的调用模式，程序将等待结果返回。
    """
    print("--- 1. invoke() (同步调用) ---")
    topic = "Runnable 协议"

    # 调用链条
    result = chain.invoke({"topic": topic})

    print(f"输入: {topic}")
    print(f"输出: {result}")
    print("-" * 30)

## 模式 2: 异步调用 (ainvoke)
async def example_ainvoke():
    """
    异步调用：在 Python 异步代码 (asyncio) 中使用，不阻塞主线程。
    """
    print("--- 2. ainvoke() (异步调用) ---")
    topic = "Poetry 包管理器"

    # 异步调用链条
    result = await chain.ainvoke({"topic": topic})

    print(f"输入: {topic}")
    print(f"输出: {result}")
    print("-" * 30)

## 模式 3: 批量调用 (batch)
async def example_batch():
    """
    批量调用：一次性传入多个输入，模型底层会优化请求，以提高效率。
    """
    print("--- 3. batch() (批量调用) ---")
    topics = [
        {"topic": "HTTP 协议"},
        {"topic": "SSH 密钥"},
        {"topic": "Python 虚拟环境"},
    ]

    # 批量调用链条
    results = chain.batch(topics)

    print(f"输入数量: {len(topics)}")
    print("批量输出:")
    for i, res in enumerate(results):
        print(f"  {i+1}. {res[:50]}...") # 打印结果前50个字符
    print("-" * 30)

## 模式 4: 同步流式调用 (stream)
async def example_stream():
    """
    同步流式调用：实时获取模型的输出块，适合用于构建聊天界面。
    """
    print("--- 4. stream() (同步流式调用) ---")
    topic = "LLM 大语言模型"

    print(f"输入: {topic}")
    print("流式输出 (实时): ")
    print("  ", end="")

    # 流式调用链条
    for chunk in chain.stream({"topic": topic}):
        # 这里的 end = "" ;抑制换行,确保文字在同一行 连续 出现。
        # flush=True; 强制刷新缓冲区,确保文字是 实时 逐字显示，而不是延迟出现。
        print(chunk, end="", flush=True) # 实时打印输出块

    print("\n[流式传输结束]")
    print("-" * 30)

## 模式 5: 异步流式调用 (astream)
async def example_astream():
    """
    异步流式调用：在异步环境中实时获取模型的输出块，常用于高并发的 Web 服务。
    """
    print("--- 5. astream() (异步流式调用) ---")
    topic = "Python asyncio"

    print(f"输入: {topic}")
    print("异步流式输出 (实时): ")
    print("  ", end="")

    # 异步流式调用链条
    async for chunk in chain.astream({"topic": topic}):
        print(chunk, end="", flush=True) # 实时打印输出块

    print("\n[异步流式传输结束]")
    print("-" * 30)


# 运行所有示例
async def main():
    await example_invoke()
    await example_ainvoke()
    await example_batch()
    await example_stream()
    await example_astream()

# 调用主函数
await main()
```

# 4. 提示词模版

1. 字符串提示模板 (StringPromptTemplate)
这是最基础、最通用的模板类型。它将输入变量直接替换到模板字符串的占位符中。它适用于简单的、不需要多轮对话或角色设定的场景。

🌟 核心类：PromptTemplate
特点： 模板完全由一个 Python 字符串定义，使用 f-string 风格的 {variable} 占位符。

用途： 问答、摘要、翻译等单次交互任务。

💻 示例

```python
from langchain_core.prompts import PromptTemplate

# 定义模板
template = "请用 {language} 语言，以一个 {role} 的身份，总结以下内容：\n{text}"

# 创建实例
prompt = PromptTemplate(
    input_variables=["language", "role", "text"],
    template=template,
)

# 格式化/调用
formatted_prompt = prompt.format(
    language="中文",
    role="资深程序员",
    text="LangChain 是一个用于开发由 LLM 驱动应用的框架。"
)

print(formatted_prompt)
```

2. 聊天提示模板 (ChatPromptTemplate)
这是最常用、推荐使用 的模板类型，因为它与现代的聊天模型（如 GPT-4, Claude, DeepSeek）的原生 API 设计相匹配。聊天模型 API 接收的不是一个大字符串，而是一个结构化的消息列表。

🌟 核心类：ChatPromptTemplate
特点： 它基于 角色 (System, Human, AI) 来构建提示。这能更好地指导模型行为，并精确地模拟对话。

用途： 所有的聊天应用、Agent 代理、复杂指令设置。

💻 示例

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage, HumanMessage

# 方式一：使用 from_messages 方法 (推荐)
chat_prompt = ChatPromptTemplate.from_messages([
    # 系统消息：设置模型人格和行为
    ("system", "你是一个专业的历史学家，总是提供清晰、简洁的答案。"),
    # 用户消息：接收动态输入变量
    ("human", "请解释 {topic} 的主要历史意义。")
])

# 格式化/调用
formatted_messages = chat_prompt.format_messages(topic="工业革命")

# formatted_messages 是一个 BaseMessage 对象列表，可以直接传给 Chat Model
print(formatted_messages)
```

3. 消息列表提示模板 (MessagesPlaceholder)
这种模板本身不是一个完整的提示，而是用于在 ChatPromptTemplate 中插入 动态生成的消息列表。它对于处理 聊天历史记录 或 Agent 的思考过程 至关重要。

🌟 核心类：MessagesPlaceholder
特点： 它允许您在聊天模板中，指定一个位置来插入另一个消息列表（例如，由 memory 模块提供的历史消息）。

用途： 为模型提供完整的对话历史，实现上下文感知（Context-aware）的聊天。

💻 示例

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# 假设这是您从 Memory 中获取的历史消息
history = [
    HumanMessage(content="我的名字是小明。"),
    SystemMessage(content="好的，小明，记住了。")
]

# 定义一个包含占位符的 Chat 模板
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个记忆力超群的聊天机器人。"),
    # 🌟 占位符：用于插入聊天历史
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "我的名字是什么？")
])

# 格式化/调用
# 必须为 chat_history 变量提供一个消息列表
formatted_messages = chat_prompt.format_messages(chat_history=history)

print(formatted_messages)
```

4. FewShotPromptTemplate。

这个模板本身不直接定义提示的最终结构，而是作为一种 包装器（Wrapper），它的核心作用是：

将一组高质量的“输入-预期输出”示例（Few-shot Examples）动态地插入到您定义的基础提示模板中。

🌟 FewShotPromptTemplate 详解
它通过组合以下三个核心元素来工作：

1. 基础提示模板 (Prefix and Suffix)
    这是围绕示例的固定文本部分。

    前缀 (Prefix)： 放在示例之前，通常用来给 LLM 提供指令和背景。

    后缀 (Suffix)： 放在示例之后，通常是用户提出的实际问题（需要 LLM 回答的部分）。

2. 示例列表 (Examples)
    这是一个包含多个字典的列表，每个字典都代表一个高质量的输入和输出的配对。

    例如：[{"input": "cat", "output": "meow"}, {"input": "dog", "output": "woof"}]

3. 示例格式化器 (Example Selector / Example Prompt)
    定义每个示例应该如何被格式化成 LLM 提示的一部分，以及如何选择这些示例。

💻 核心用法和结构
FewShotPromptTemplate 通常与一个基础的 PromptTemplate 结合使用。

```python
from langchain_core.prompts import (
    FewShotPromptTemplate,
    PromptTemplate,
)

# --- 1. 定义 Few-shot 示例 ---
# 示例：将 IT 概念翻译成大白话
examples = [
    {
        "input": "什么是 RESTful API？",
        "output": "可以理解为餐厅的服务员，你告诉他你要什么菜（请求），他去后厨（服务器）拿给你，每次都是独立的交易。",
    },
    {
        "input": "什么是 CI/CD？",
        "output": "可以理解为工厂的自动化生产线，一旦工程师改动了代码，系统就自动进行测试和部署，无需人工干预。",
    },
    # ... 更多示例 ...
]

# --- 2. 定义单个示例的格式化模板 ---
# 告诉 LangChain 应该如何打印每个 'input' 和 'output' 示例
example_template = "Q: {input}\nA: {output}"
example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template=example_template,
)

# --- 3. 组合成 FewShotPromptTemplate ---
few_shot_prompt = FewShotPromptTemplate(
    # 用于格式化单个示例的模板
    example_prompt=example_prompt,
    # 示例数据
    examples=examples,
    # 整个提示的前缀 (指令)
    prefix="你是专业的 IT 概念转译器，请使用生动的比喻来解释复杂的概念。下面是一些范例：\n",
    # 整个提示的后缀 (用户提问) - 这里需要传入 'input' 变量
    suffix="\nQ: {input}\nA:",
    # 提示中所需的输入变量（即后缀中使用的变量）
    input_variables=["input"],
)

# --- 4. 格式化并调用 ---
final_prompt = few_shot_prompt.format(input="什么是 Docker 容器？")

print(final_prompt)
```
