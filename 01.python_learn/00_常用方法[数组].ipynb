# 1.reversed()
倒序reversed() 必须用for 循环

```python
list = [1, 2, 3, 4, 5]
relist = [i for i in reversed(list)]
print(relist)
```

# 2.next() 和 iter()
  next()：从一个可迭代流程里，拿“下一个结果”, 从0 开始
 iter() 变成可迭代对象，例如for循环

```python
list = iter([1, 2, 3, 4, 5])
print(next(list))
print(next(list))

list2 = [1, 2, 3, 4, 5]
print(next(i for i in list2))
```

# 3.查找 find， 匹配到就停止
   采用next()的方式

```python
list = [1, 2, 3, 4, 5]

targer = next(item for item in list if item > 3)

print(targer)
```
